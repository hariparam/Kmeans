```python
import gzip
import pandas as pd
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
import timeit
from sklearn.cluster import KMeans, MiniBatchKMeans
import numpy as np
import os
%matplotlib inline
```

```python
reptn=15
n_clusters=[5, 10, 25, 50, 100, 200, 500 ]
cluster_size=[str(i) for i in n_clusters]
max_iterations=[5, 10, 25, 50, 100, 300]
init_methods=['k-means++', 'random']
#n_clusters=20
col_list=['max_iterations', 'init_method', 'cluster_size', 'values','avg', 'max', 'min']
col_list= col_list
```

```python
def addrow(max_iterations,init_method,cluster_size):
    X=dict()
    X['max_iterations']=max_iterations
    X['init_method']=init_method
    X['cluster_size']=cluster_size
    X['values']=[]
    X['avg']=0
    X['max']=0
    X['min']=0
    return X
```

```python
datafiles=[]
for i in os.listdir('.'):
    if 'ydata' in i:
        datafiles.append(i)
datafiles
```

```python
for files in datafiles:
    f=gzip.open(files)
    values=[]
#     for i in range(10):
#         print(f.readline())
    for line in tqdm(f):
        spl=line.split("|")
        values.append([float(x.split(':')[1]) for x in spl[1].split(' ') if ':' in x])
    print(files,len(values))
    col_list=['feature_'+str(i) for i in range(len(values[0]))]
    clicks=pd.DataFrame(values, columns=col_list)    
```

```python

for indx, files in enumerate(datafiles):
    f=gzip.open(files)
    values=[]
#     for i in range(10):
#         print(f.readline())
    start_time = timeit.default_timer()
    for line in (f):
        spl=line.split("|")
        values.append([float(x.split(':')[1]) for x in spl[1].split(' ') if ':' in x])
    elapsed = timeit.default_timer() - start_time
    print(files,len(values), elapsed)
    col_list=['feature_'+str(i) for i in range(len(values[0]))]
    clicks=pd.DataFrame(values, columns=col_list)
    clicks.to_csv("clicks_"+str(indx)+".csv")
```

```python
# timing_mini=dict()
# error_mini=dict()
batch=100000 
for core in range(reptn):
    for i in n_clusters:
        for k in init_methods:
            print(k)
            centroids=[]
            for j in max_iterations:
                label=str(i)+'_'+str(k)+'_'+str(j)
                if (label not in timing_mini.keys()):
                    timing_mini[label]=[]
                    error_mini[label]=[]
                start_time = timeit.default_timer()
                if (j==5):
                    kmeans = MiniBatchKMeans(n_clusters=i,init=k,max_iter=j,batch_size=batch)
                else:
                    kmeans = MiniBatchKMeans(n_clusters=i,init=centroids,max_iter=j,batch_size=batch)
                for f in df:
                    clicks=pd.read_csv(f, index_col=None)
                    clicks=clicks.drop(columns=['Unnamed: 0'])
                    for b in (range(0,len(clicks), batch)):
                        kmeans.partial_fit(clicks[b: min(len(clicks), b+batch)])
                elapsed = timeit.default_timer() - start_time
                print(core,"cluster_size :", i, "iterations", j, "time :", elapsed, "quant_error :", kmeans.inertia_)
                centroids=kmeans.cluster_centers_
                timing_mini[label].append(elapsed)
                error_mini[label].append(kmeans.inertia_)
```

```python
timing_results2=pd.DataFrame(columns=col_list)
quantizationerror_results2=pd.DataFrame(columns=col_list)
for i in n_clusters:
    for j in max_iterations:
        for k in init_methods:
            label=str(i)+'_'+str(k)+'_'+str(j)
            X=addrow(j,k,i)
            X['values']=timing_mini[label]
            X['avg']=sum(timing_mini[label])/len(timing_mini[label])
            X['max']=max(timing_mini[label])
            X['min']=min(timing_mini[label])
            timing_results2.loc[len(timing_results2)]=X
            X['values']=error_mini[label]
            X['avg']=sum(error_mini[label])/len(error_mini[label])
            X['max']=max(error_mini[label])
            X['min']=min(error_mini[label])
            quantizationerror_results2.loc[len(quantizationerror_results2)]=X
```
