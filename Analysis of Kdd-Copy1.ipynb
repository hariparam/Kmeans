```python
def kplusplus_init(X, K):
    X=np.array(X)
    C = [X[np.array(random.sample(range(len(X)), 1))].flatten()]
    #C=[(X[0])]
    for k in range(1, K):
        start_time = timeit.default_timer()
        D2 = np.array([min([np.inner(c-x,c-x) for c in C]) for x in X])
        probs = D2/D2.sum()
        cumprobs = probs.cumsum()
        r = np.random.rand()
        for j,p in enumerate(cumprobs):
            if r < p:
                i = j
                break
        elapsed = timeit.default_timer() - start_time
        C.append(X[i])
        print(k,i, elapsed)
    return np.array(C)
```
